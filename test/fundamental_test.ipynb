```python
BS = pd.read_sql(f"aapl_balance_sheet_annual", engine)
print(BS.columns)
BS
```

```python
PL = pd.read_sql(f"aapl_financials_annual", engine)
print(PL.columns)
PL
```

```python
GrossProfitMargin = PL['Gross Profit']/PL['Total Revenue']
OperatingMargin = PL['Operating Income']/PL['Total Revenue']
NetProfitMargin = PL['Net Income']/PL['Total Revenue']
EBITDAMargin = PL['EBITDA']/PL['Total Revenue']
ROA = PL['Net Income']/BS['Total Assets']
ROE = PL['Net Income']/BS['Stockholders Equity']
tax_rate = PL['Tax Provision']/PL['Pretax Income']
NOPAT = PL['EBIT'] * (1 - tax_rate)
ROIC = NOPAT/BS['Invested Capital']

print(GrossProfitMargin)
print(OperatingMargin)
print(NetProfitMargin)
print(ROA)
print(ROE)
print(tax_rate)
print(NOPAT)
print(ROIC)
```

```python
# solvency & liquidity
CurrentRatio = BS['Current Assets']/BS['Current Liabilities']
QuickRatio = (BS['Current Assets'] - BS['Inventory'])/BS['Current Liabilities']
DebtToEquityRatio = BS['Total Liabilities Net Minority Interest']/BS['Stockholders Equity']
EquityRatio = BS['Stockholders Equity']/BS['Total Assets']

print(CurrentRatio)
print(QuickRatio)
print(DebtToEquityRatio)
print(EquityRatio)
```
