```python
exp1 = 1
exp2 = 2
std1 = 1
std2 = 2
cor = 1/2

var1 = std1**2
var2 = std2**2
cov = std1*std2*cor

A = var1 + var2 - 2*cov
exp0 = (exp1*var1 + exp2*var2 - (exp1 + exp2)*cov)/A
var0 = (var1*var2 + cov**2)/A
gamma = A/(exp1 - exp2)**2
```

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_portfolio(exp1, exp2, std1, std2, cor, color="black"):
    # define parameter
    var1 = std1**2
    var2 = std2**2
    cov = std1*std2*cor

    A = var1 + var2 - 2*cov
    exp0 = (exp1*var2 + exp2*var1 - (exp1 + exp2)*cov)/A
    var0 = (var1*var2 - cov**2)/A
    gamma = A/((exp1 - exp2)**2)

    # xとyの範囲を定義
    x = np.linspace(0, 2**4, 2**12)
    y = np.linspace(0, 2**4, 2**12)

    # グリッドを作成
    X, Y = np.meshgrid(x, y)

    # 双曲線の方程式: x^2 - y^2 = 1
    equation = X**2 - gamma*((Y - exp0)**2)

    # 等高線プロット
    # plt.figure(figsize=(8, 8))
    plt.scatter(std1, exp1, color='blue')
    plt.scatter(std2, exp2, color='orange')
    plt.contour(X, Y, equation, [var0], colors=color) # 方程式 = 1となる線を引く
```

```python
# exp1 < exp2, std1 < std2
# cor < std1/std2
plot_portfolio(3, 5, 3, 5, 1/4)
# cor > std1/std2
plot_portfolio(3, 5, 3, 5, 3/4)
# cor = std1/std2
plot_portfolio(3, 5, 3, 5, 3/5)
# cor = 1
plot_portfolio(3, 5, 3, 5, 1)
# cor = -1
plot_portfolio(3, 5, 3, 5, -1)

plt.title('portfolio (exp1 < exp2, std1 < std2)')
plt.xlabel('std')
plt.ylabel('exp')
plt.grid(True)
plt.savefig("image1.png")
```

```python
# exp1 > exp2, std1 < std2
# cor < std1/std2
plot_portfolio(5, 3, 3, 5, 1/4)
# cor > std1/std2
plot_portfolio(5, 3, 3, 5, 3/4)
# cor = std1/std2
plot_portfolio(5, 3, 3, 5, 3/5)
# cor = 1
plot_portfolio(5, 3, 3, 5, 1)
# cor = -1
plot_portfolio(5, 3, 3, 5, -1)

plt.title('portfolio (exp1 > exp2, std1 < std2)')
plt.xlabel('std')
plt.ylabel('exp')
plt.grid(True)
plt.savefig("image2.png")
```
